# A basic guide to creating plots in Matplotlib

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
ax.plot(x, y, linestyle='-', color='red')

ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
ax.set_xlim([0, 6])
ax.set_ylim([0, 12])
plt.show()
```

## Use a colour palette

```python
x = [1,2,3,4,5,6,7]
y = [4,5,6,7,1,2,8]
z = [1,2,3,4,5,6,7]
fig, ax = plt.subplots()
ax.scatter(x, y, c=z, cmap='rainbow')
ax.set_title("Using colour palette's to customize a plot")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")

plt.show()
```

## Create subplots

```python
x=[1,2,3,4,5,6,7,8,9]
y=[1,2,3,4,5,6,7,8,9]

fig, ax = plt.subplots(2, 2)
fig.suptitle("Subplots example")
ax[0, 0].plot(x, y, linestyle='-', color='red')
ax[0, 0].title.set_text('First Plot')
ax[0, 1].plot(x, y, linestyle='-', color='blue')
ax[0, 1].title.set_text('Second Plot')
ax[1, 0].plot(x, y, linestyle='-', color='green')
ax[1, 0].title.set_text('Third Plot')
ax[1, 1].plot(x, y, linestyle='-', color='yellow')
ax[1, 1].title.set_text('Fourth Plot')
fig.subplots_adjust(hspace=0.4, wspace=0.4)
plt.show()
```
